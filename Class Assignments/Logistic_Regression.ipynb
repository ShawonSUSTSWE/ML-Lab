```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import copy
import math
from matplotlib.patches import FancyArrowPatch
from sklearn.linear_model import LogisticRegression
from ipywidgets import Output
from google.colab import drive
drive.mount('/content/drive')
```

```python
data = pd.read_csv('/content/drive/MyDrive/ML Lab/data.txt', names = ['X0', 'X1', 'y'])
data.head()
```

```python
x = data[['X0', 'X1']].values
y = data['y'].values

pos_label = "y=1"
neg_label = "y=0"

fig, ax = plt.subplots(1, 1, figsize=(6, 6))

pos = y == 1
neg = y == 0

ax.scatter(x[pos, 0], x[pos, 1], marker='x', s=80, c='red', label=pos_label)
ax.scatter(x[neg, 0], x[neg, 1], marker='o', s=80, label=neg_label, facecolors='none', edgecolors='green', lw=2)
ax.legend(loc='best')
```

```python
lreg_model = LogisticRegression()
lreg_model.fit(x, y)
```

```python
min_x0, max_x0 = x[:, 0].min() - 1, x[:, 0].max() + 1
fig, ax = plt.subplots(1, 1, figsize=(8, 6))

ax.scatter(x[pos, 0], x[pos, 1], marker='x', s=80, c='red', label=pos_label)
ax.scatter(x[neg, 0], x[neg, 1], marker='o', s=80, label=neg_label, facecolors='none', edgecolors='green', lw=2)
ax.legend(loc='upper right')

x0_values = np.linspace(min_x0, max_x0, 100)
x1_values = -(lreg_model.intercept_ + lreg_model.coef_[0, 0] * x0_values) / lreg_model.coef_[0, 1]

ax.plot(x0_values, x1_values, c='green', lw=1)
plt.show()
```
